```python
import matplotlib.pyplot as plt 
%matplotlib inline #to get the figures
```

# pie chart

```python
areas = ['Marketing','sales','development','hr','customer','support']
budget = [2.4,1.3,1.4,10,1.5,12]

plt.pie(budget,labels = areas,autopct = "%0.2f%%")
#to remove technical data use
plt.show()
```

```python
import numpy as np 
x = np.arange(30,50)
y = np.arange(23,43)
plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('salary')
plt.title('Age vs Salary')
plt.show()
```

# line chart

```python
product_id = [1,2,3,4,5,6,7,8]
online_prices = [224,224,23,243,534,65,34,54]
offline_prices = [400,200,324,234,432,54,43,234]
chor_bazar = [200,200,30,40,50,30,20,60]

plt.plot(product_id,online_prices,color = 'red',marker='D',markersize = 10,label = "online_prices")#2d plot
plt.plot(product_id,offline_prices,color = 'yellow',markersize = 10,label = "offline_prices")
plt.plot(product_id,chor_bazar,color = 'green',marker = "+",markersize = 10,label= "chor_bazar_prices")

plt.xlabel('product')
plt.ylabel('prices')
plt.title('product and price comparison')
plt.legend()# the good thing about legend is it finds the space to adjust itself in the graph.

plt.show()
```

# bar graph

```python
year = [2012,2013,2014,2015,2016,2017,2018]
price = [64000,69000,75000,34000,60000,70000,120000]
plt.bar(year,price,color = 'orange')
plt.xlabel('Year')
plt.ylabel('price')
plt.title('year by year bike price')
plt.show()
```

```python
plt.barh(year,price,color = 'red')
plt.show()
```
